```python
import random
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Step 1: Generate 100 random numbers between 0 and 1
random.seed(21)
data = [round(random.random(), 3) for _ in range(100)]

# Step 2: Label the data: 1 if <= 0.5, else 2
labels = [1 if d <= 0.5 else 2 for d in data]

# Step 3: Split into training and test sets (first 50 for training, rest for testing)
X_train = np.array(data[:50]).reshape(-1, 1)
y_train = np.array(labels[:50])
X_test = np.array(data[50:]).reshape(-1, 1)
y_test = np.array(labels[50:])

# Step 4: Train and evaluate KNN with k=1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Accuracy (k=1):", knn.score(X_test, y_test))

# Step 5: Plot accuracy for different k values (test data)
k_values = [1, 2, 3, 4, 5, 20, 30]
test_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    test_scores.append(knn.score(X_test, y_test))

plt.plot(k_values, test_scores, marker='o')
plt.title('KNN Accuracy on Test Data')
plt.xlabel('K value')
plt.ylabel('Accuracy')
plt.show()

# Step 6: Plot accuracy for different k values (train data)
train_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_scores.append(knn.score(X_train, y_train))

plt.plot(k_values, train_scores, marker='o', color='orange')
plt.title('KNN Accuracy on Train Data')
plt.xlabel('K value')
plt.ylabel('Accuracy')
plt.show()
```
