```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd

df=pd.read_csv(r'BostonHousing.csv')

x=df[['RM']]
y=df['MEDV']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

print("Intercept:",model.intercept_)
print("Coefficent:",model.coef_[0])

plt.scatter(x_test,y_test,color='blue',label='Actual')
plt.plot(x_test,y_pred,color='red',label='Predicted')
plt.xlabel("Avg number of rooms")
plt.ylabel("House Price in (1000$)")
plt.title("Linear Regression: Boston Housing")
plt.legend()
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score, mean_squared_error

# --- Linear Regression on Boston Housing ---
data = pd.read_csv("BostonHousing.csv")
x = data[['RM']]
y = data['MEDV']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print("Linear Regression:")
print("R2:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))

plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_pred, color='red', label='Predicted')
plt.xlabel("Rooms")
plt.ylabel("Price")
plt.title("Boston Housing")
plt.legend()
plt.show()

# --- Polynomial Regression on Auto MPG ---
data = pd.read_csv("auto-mpg.csv")
data = data[data['horsepower'] != '?']
data['horsepower'] = data['horsepower'].astype(float)

x = data[['horsepower']]
y = data['mpg']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = make_pipeline(PolynomialFeatures(2), LinearRegression())
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print("\nPolynomial Regression:")
print("R2:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))

x_line = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
x_line_df = pd.DataFrame(x_line, columns=['horsepower'])
y_line = model.predict(x_line_df)

plt.scatter(x, y, alpha=0.3, label='Actual')
plt.plot(x_line_df, y_line, color='red', label='Fit')
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.title("Auto MPG")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error

data=pd.read_csv(r'BostonHousing.csv')

x=data[['RM']]
y=data['MEDV']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

print("Linear Regression:\n")
print("R2 score:",r2_score(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))

plt.scatter(x_test,y_test,color='blue',label='Actual')
plt.plot(x_test,y_pred,color='red',label='Predicted')
plt.title("Boston Housing")
plt.xlabel('Rooms')
plt.ylabel('Prices')
plt.legend()
plt.show()
```
